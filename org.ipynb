```python
from matplotlib import colors
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.style
import matplotlib as mpl
import matplotlib.animation as pltan

org= pd.read_csv('C:\\Users\\HP PAVILION GAMING\\Desktop\\new19_9\\organizations.csv')
male=org.loc[org['gender'] == 'male'].count()[0]
female=org.loc[org['gender'] == 'female'].count()[0]
orga=org.loc[org['gender'] == 'org'].count()[0]
colors1 = ['#FFFF14', '#00FF00', '#0000FF']
labels = ['male', 'female', 'organizations (all in the field of peace)']
explode=(.1, .1, 0.1)
plt.pie([male, female, orga], labels=labels, colors=colors1, autopct='%.2f %%', pctdistance=0.8, explode=explode, wedgeprops = {"edgecolor" : "#650021",
                      'linewidth': 1, 'antialiased': True})
plt.title('Procentage of male, female and organizations by categories', color='blue')
plt.show()
```
