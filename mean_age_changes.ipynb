```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.axes_grid1 import Divider, Size
df=pd.read_csv('egzel2.csv')
df.head()
```

```python
dff=df[['category','win_year', 'age_winners']]
dff.head()
```

```python
dfsort = dff.sort_values('win_year')
dfsort
```

```python
dfch=dfsort.loc[dfsort['category']=='chemistry']
dfch.head(2)
```

```python
from mpl_toolkits.axes_grid1 import Divider, Size
x= dfch.win_year
y= dfch.age_winners

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

f = plt.figure()
f.set_figwidth(14)
f.set_figheight(10)
#ax.grid(True)
plt.plot(x, y, color='blue')
plt.title('Ages of Nobel prize winners in Chemistry during 1900-2020', fontdict = font1, loc = 'center')
plt.legend(['chemistry'], bbox_to_anchor=(1,1), loc="upper left")
plt.ylabel('Ages', fontdict = font2, loc = 'center')
plt.xlabel('Years', fontdict = font2, loc = 'center')
plt.show()
```

```python
# creating of list for decades
a=[]
def adding():
    for x in dfli.win_year:
        if 1900 <= x >= 1910:
            a.append('1900-1910')
        if  1911 <= x >= 1920:
            a.append('1911-1920')
        if 1921 <= x >= 1930:
            a.append('1921-1930')
        if 1931 <= x >= 1940:
            a.append('1931-1940')
        if 1941 <= x >= 1950:
            a.append('1941-1950') 
        if 1951 <= x >= 1960:
            a.append('1951-1960')
        if 1961 <= x >= 1970:
            a.append('1961-1970')
        if 1971 <= x >= 1980:
            a.append('1971-1980')
        if 1981 <= x >= 1990:
            a.append('1981-1990')
        if 1991 <= x >= 2000:
            a.append('1991-2000')
        if 2001 <= x >= 2010:
            a.append('2001-2010')
        if 2011 <= x >= 2020:
            a.append('2011-2020')
adding() 

print (a)
```

```python
dd=pd.read_csv('decades.csv')
dd
```

```python
chem=dd.loc[dd['category'] == 'chemistry']
phy=dd.loc[dd['category'] == 'physics']
med=dd.loc[dd['category'] == 'medicine']
eco=dd.loc[dd['category'] == 'economics']
lit=dd.loc[dd['category'] == 'literature']
pea=dd.loc[dd['category'] == 'peace']
```

```python
chem_agg = chem.groupby(by=['decades','category']).age_winners.mean().reset_index()
phy_agg = phy.groupby(by=['decades','category']).age_winners.mean().reset_index()
med_agg = med.groupby(by=['decades','category']).age_winners.mean().reset_index()
eco_agg = eco.groupby(by=['decades','category']).age_winners.mean().reset_index()
lit_agg = lit.groupby(by=['decades','category']).age_winners.mean().reset_index()
pea_agg = pea.groupby(by=['decades','category']).age_winners.mean().reset_index()
```

```python
from typing import Sized
import pandas as pd
import datetime 
import numpy as nm
from matplotlib import pyplot as plt

x1= chem_agg.decades
x2= phy_agg.decades
x3= med_agg.decades
x4= eco_agg.decades
x5= lit_agg.decades
x6= pea_agg.decades


y1=chem_agg.age_winners
y2=phy_agg.age_winners
y3= med_agg.age_winners
y4=eco_agg.age_winners
y5=lit_agg.age_winners
y6=pea_agg.age_winners

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

f = plt.figure()
f.set_figwidth(14)
f.set_figheight(10)
#ax.grid(True)
plt.plot(x1, y1, color='blue')
plt.plot(x2, y2, color='red')
plt.plot(x3, y3, color='green')
plt.plot(x4, y4, color='yellow')
plt.plot(x5, y5, color='pink')
plt.plot(x6, y6, color='black')

plt.title('Ages of Nobel prize winners in all categories in decades during 1900-2020', fontdict = font1, loc = 'center')
plt.legend(['Chemistry', 'Physics', 'Medicine', 'Economics', 'Literature', 'Peace'], bbox_to_anchor=(1,1), loc="upper left")
plt.ylabel('Ages', fontdict = font2, loc = 'center')
plt.xlabel('Decades', fontdict = font2, loc = 'center')
plt.xticks(rotation=90)
#ax.grid(True)
plt.show()
```

```python
chem_agg = chem.groupby(by=['twentees','category']).age_winners.mean().reset_index()
phy_agg = phy.groupby(by=['twentees','category']).age_winners.mean().reset_index()
med_agg = med.groupby(by=['twentees','category']).age_winners.mean().reset_index()
eco_agg = eco.groupby(by=['twentees','category']).age_winners.mean().reset_index()
lit_agg = lit.groupby(by=['twentees','category']).age_winners.mean().reset_index()
pea_agg = pea.groupby(by=['twentees','category']).age_winners.mean().reset_index()
```

```python
x1= chem_agg.twentees
x2= phy_agg.twentees
x3= med_agg.twentees
x4= eco_agg.twentees
x5= lit_agg.twentees
x6= pea_agg.twentees


y1=chem_agg.age_winners
y2=phy_agg.age_winners
y3= med_agg.age_winners
y4=eco_agg.age_winners
y5=lit_agg.age_winners
y6=pea_agg.age_winners

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

f = plt.figure()
f.set_figwidth(14)
f.set_figheight(10)
#ax.grid(True)
plt.plot(x1, y1, color='blue', linestyle = ':',
        linewidth = 6,)
plt.plot(x2, y2, color='red', linestyle = '--',
        linewidth = 3)
plt.plot(x3, y3, color='green', marker = 'X', markersize = 20,  markeredgecolor = 'teal', 
            markerfacecolor = 'greenyellow', markeredgewidth = 3)
plt.plot(x4, y4, color='yellow', linewidth = 6)
plt.plot(x5, y5, color='pink', marker = '^', markeredgecolor = 'darkgreen', markersize = 15, linewidth = 6)
plt.plot(x6, y6, color='cyan', marker = 'o',
        markersize = 13, markerfacecolor = 'cyan',
        markeredgecolor = 'darkgreen', markeredgewidth = 3, linewidth = 6)

plt.title('Avarage ages of Nobel prize winners in all categories in twentees during 1900-2020', fontdict = font1, loc = 'center')
plt.legend(['Chemistry', 'Physics', 'Medicine', 'Economics', 'Literature', 'Peace'], bbox_to_anchor=(1,1), loc="upper left")
plt.ylabel('Ages', fontdict = font2, loc = 'center')
plt.xlabel('Twentees', fontdict = font2, loc = 'center')
plt.xticks(rotation=90)
#ax.grid(True)
plt.show()
```
