```python
# I prepared csv file with counting  frequency all zodiacs in all categories for further visualization 
m=pd.read_csv('ham_mon.csv')
m
```

```python
#polar chart of frequency of month zodiacs in the category of chemistry
from matplotlib import pyplot as plt
import numpy as np

freq = m['frequencych']
zodiacs = ['aquarious','aries','cancer', 'capricorn','gemini','leo','libra', 'pisces','sagittarius',  'scorpio', 'taurus',    
              'virgo']


fr = np.array(freq)
fig = plt.figure(figsize=(8,8))
colors=np.array(['lime', 'red', 'cyan', 'bisque', 'violet', 'yellow', 'grey', 
                 'lightcoral', 'lightsteelblue', 'pink', 'orange', 'blue'])

N = 12
theta=np.arange(0,2*np.pi,2*np.pi/N)
width = (2*np.pi)/N *0.9
bottom = True
bottom=0
rotations = np.rad2deg(theta)
ax = fig.add_axes([0.1, 0.1, 0.75, 0.75], polar=True)
bars = ax.bar(theta, freq, width=width, bottom=bottom, color=colors)
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'black','size':15}
plt.title("Frequency of month zodiacs in the category of chemistry", fontdict = font1)
for x, bar, rotation, label in zip(theta, bars, rotations, zodiacs):
    lab = ax.text(x,bottom+bar.get_height() , label, fontdict = font2,
             ha='right', va='center', rotation=rotation, rotation_mode="anchor")  
plt.show()
```
