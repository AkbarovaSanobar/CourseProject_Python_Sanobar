```python
# month zodiak in the field of chemistry
chemistry_month = mzc[mzc.category == "chemistry"]
chemistry_month
```

```python
# counting of number of month zodiak in the field of chemistry
ch_m=chemistry_month['zod_month'].value_counts(normalize=True)
ch_m
```

```python
#bar chart of month zodiac in chemistry
my_plot = ch_m.plot(kind='bar', figsize=(9, 7))
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
my_plot.set_xlabel("Zodiacs of month", fontdict = font2)
my_plot.set_ylabel("Frequency occurence", fontdict = font2)
plt.title('Frequency of month zodiacs in the category of chemistry', fontdict = font1, loc = 'center')
plt.tight_layout()
```

```python
ch1 = pd.read_csv('ch_month.csv')
ch1.rename(columns= {'Unnamed: 0':'zod_m_ch', 'zod_month':'frequencych'}, inplace=True)
ph1 = pd.read_csv('ph_month.csv')
ph1.rename(columns= {'Unnamed: 0':'zod_m_ph', 'zod_month':'frequencyph'}, inplace=True)
med1 = pd.read_csv('me_month.csv')
med1.rename(columns= {'Unnamed: 0':'zod_m_med', 'zod_month':'frequencymed'}, inplace=True)
lit1 = pd.read_csv('li_month.csv')
lit1.rename(columns= {'Unnamed: 0':'zod_m_lit', 'zod_month':'frequencylit'}, inplace=True)
ec1 = pd.read_csv('ec_month.csv')
ec1.rename(columns= {'Unnamed: 0':'zod_m_ec', 'zod_month':'frequencyec'}, inplace=True)
pe1 = pd.read_csv('pe_month.csv')
pe1.rename(columns= {'Unnamed: 0':'zod_m_pe', 'zod_month':'frequencype'}, inplace=True)
```

```python
ch1 = ch1.sort_values('zod_m_ch')
ph1 = ph1.sort_values('zod_m_ph')
med1 = med1.sort_values('zod_m_med')
lit1 = lit1.sort_values('zod_m_lit')
ec1 = ec1.sort_values('zod_m_ec')
pe1 = pe1.sort_values('zod_m_pe')
ch1.to_csv('ch2_m_sort.csv', index=False)
ph1.to_csv('ph2_m_sort.csv', index=False)
med1.to_csv('med2_m_sort.csv', index=False)
lit1.to_csv('lit2_m_sort.csv', index=False)
ec1.to_csv('ec2_m_sort.csv', index=False)
pe1.to_csv('pe2_m_sort.csv', index=False)
```
