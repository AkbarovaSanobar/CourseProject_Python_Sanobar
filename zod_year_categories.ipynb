```python
from typing import Sized
import pandas as pd
import datetime 
import numpy as nm
from matplotlib import pyplot as plt

hamma= pd.read_csv('C:\\Users\\HP PAVILION GAMING\\OneDrive\\Рабочий стол\\new19_9\\birlashgan.csv')
print(hamma)

plt.plot(hamma.year_zodiac, hamma.eco_frequency, color='blue')
plt.plot(hamma.year_zodiac, hamma.lit_frequency, color='red')
plt.plot(hamma.year_zodiac, hamma.pe_frequency, color = 'green')
plt.plot(hamma.year_zodiac, hamma.med_frequency, color = "yellow")
plt.plot(hamma.year_zodiac, hamma.ph_frequency, color = "tab:brown")
plt.plot(hamma.year_zodiac, hamma.ch_frequency, color = "tab:purple")


font1 = {'family':'serif','color':'blue','size':20}
plt.title('Nobel prize winners\' zodiacs of born year and categories', fontdict = font1, loc = 'center')
plt.legend(['economy', 'literature', 'medicine', 'peace', 'physics', 'chemistry'], bbox_to_anchor=(1,1), loc="upper left")

plt.xlabel('zodiacs of year')
plt.xticks(rotation=90)
plt.ylabel('frequency')

ax = plt.subplot()
ax.grid(True)
plt.show()
```

```python
my_plot = hamma.set_index('year_zodiac').plot(kind='bar', stacked=True,title="Total frequency of winners by categories",figsize=(9, 7))
my_plot.set_xlabel("Born year zodiacs")
my_plot.set_ylabel("Frequency")
my_plot.legend(["Economy","Literature","Medicine","Peace", "Physics", "Chemistry"], bbox_to_anchor=(1,1), loc="upper left",ncol=1)
```

```python
#hamma['all_categories_sum'] = hamma.groupby(['eco_frequency', 'lit_frequency', 'pe_frequency', 'med_frequency', 'ph_frequency', 'ch_frequency']).sum()
hamma['all_categories_sum'] = hamma['eco_frequency'] + hamma['lit_frequency'] + hamma['pe_frequency']+hamma['med_frequency']+hamma['ph_frequency']+hamma['ch_frequency']
hamma.to_csv('hamma2.csv', index=False)
```
