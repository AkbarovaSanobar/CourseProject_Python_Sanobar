born yera zodiacs of winners in categories

```python
from typing import Sized
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

h= pd.read_csv('C:\\Users\\HP PAVILION GAMING\\OneDrive\\Рабочий стол\\new19_9\\my_data111.csv')
df=h[['category', 'zod_year','number']]
df
```

```python
df_agg = df.groupby(by=['category', 'zod_year']).number.sum().reset_index()
df_agg
```

```python
df_pivot = pd.pivot_table(df_agg, values='number', index='category', columns='zod_year', aggfunc=np.mean)
df_pivot
```

```python
# Grouped bar chart 
ax = df_pivot.plot(kind="bar", width=0.8)
#width = 1.5
fig = ax.get_figure()
fig.set_size_inches(16, 8)
ax.set_xlabel("Categories")
ax.set_ylabel("Frequency of year zodiacs")
ax.set_title('Frequency of year zodiac in all the categories', pad=15)
ax.legend( bbox_to_anchor=(1,1), loc="upper left",ncol=1)
for bar in ax.patches:
    bar_value = bar.get_height()
    text = f'{bar_value:,}'
    text_x = bar.get_x() + bar.get_width() / 2
    text_y = bar.get_y() + bar_value
    bar_color = bar.get_facecolor()
    ax.text(text_x, text_y, text, ha='center', va='bottom', color='black',size=10)
def change_width(ax, new_value) :
    for patch in ax.patches :
        current_width = patch.get_width()
        diff = current_width - new_value

        # we change the bar width
        patch.set_width(new_value)

        # we recenter the bar
        patch.set_x(patch.get_x() + diff * .5)

change_width(ax, .15)
ax.grid(True)
plt.show()

#fig.savefig("page_views_barplot.png")
```

```python
from typing import Sized
import pandas as pd
import datetime 
import numpy as nm
from matplotlib import pyplot as plt

hamma= pd.read_csv('C:\\Users\\HP PAVILION GAMING\\OneDrive\\Рабочий стол\\new19_9\\birlashgan.csv')
print(hamma)

plt.plot(hamma.year_zodiac, hamma.eco_frequency, color='blue')
plt.plot(hamma.year_zodiac, hamma.lit_frequency, color='red')
plt.plot(hamma.year_zodiac, hamma.pe_frequency, color = 'green')
plt.plot(hamma.year_zodiac, hamma.med_frequency, color = "yellow")
plt.plot(hamma.year_zodiac, hamma.ph_frequency, color = "tab:brown")
plt.plot(hamma.year_zodiac, hamma.ch_frequency, color = "tab:purple")


font1 = {'family':'serif','color':'blue','size':20}
plt.title('Nobel prize winners\' zodiacs of born year and categories', fontdict = font1, loc = 'center')
plt.legend(['economy', 'literature', 'medicine', 'peace', 'physics', 'chemistry'], bbox_to_anchor=(1,1), loc="upper left")

plt.xlabel('zodiacs of year')
plt.xticks(rotation=90)
plt.ylabel('frequency')

ax = plt.subplot()
ax.grid(True)
plt.show()
```
