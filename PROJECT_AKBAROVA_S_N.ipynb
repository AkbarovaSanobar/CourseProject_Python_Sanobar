```python
# creating of list for decades
a=[]
def adding():
    for x in dfli.win_year:
        if 1900 <= x >= 1910:
            a.append('1900-1910')
        if  1911 <= x >= 1920:
            a.append('1911-1920')
        if 1921 <= x >= 1930:
            a.append('1921-1930')
        if 1931 <= x >= 1940:
            a.append('1931-1940')
        if 1941 <= x >= 1950:
            a.append('1941-1950') 
        if 1951 <= x >= 1960:
            a.append('1951-1960')
        if 1961 <= x >= 1970:
            a.append('1961-1970')
        if 1971 <= x >= 1980:
            a.append('1971-1980')
        if 1981 <= x >= 1990:
            a.append('1981-1990')
        if 1991 <= x >= 2000:
            a.append('1991-2000')
        if 2001 <= x >= 2010:
            a.append('2001-2010')
        if 2011 <= x >= 2020:
            a.append('2011-2020')
adding() 

print (a)
```

```python
dd=pd.read_csv('decades.csv')
dd
```

```python
chem=dd.loc[dd['category'] == 'chemistry']
phy=dd.loc[dd['category'] == 'physics']
med=dd.loc[dd['category'] == 'medicine']
eco=dd.loc[dd['category'] == 'economics']
lit=dd.loc[dd['category'] == 'literature']
pea=dd.loc[dd['category'] == 'peace']
```

```python
chem_agg = chem.groupby(by=['decades','category']).age_winners.mean().reset_index()
phy_agg = phy.groupby(by=['decades','category']).age_winners.mean().reset_index()
med_agg = med.groupby(by=['decades','category']).age_winners.mean().reset_index()
eco_agg = eco.groupby(by=['decades','category']).age_winners.mean().reset_index()
lit_agg = lit.groupby(by=['decades','category']).age_winners.mean().reset_index()
pea_agg = pea.groupby(by=['decades','category']).age_winners.mean().reset_index()
```

```python
x1= chem_agg.decades
x2= phy_agg.decades
x3= med_agg.decades
x4= eco_agg.decades
x5= lit_agg.decades
x6= pea_agg.decades


y1=chem_agg.age_winners
y2=phy_agg.age_winners
y3= med_agg.age_winners
y4=eco_agg.age_winners
y5=lit_agg.age_winners
y6=pea_agg.age_winners

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

f = plt.figure()
f.set_figwidth(14)
f.set_figheight(10)
#ax.grid(True)
plt.plot(x1, y1, color='blue')
plt.plot(x2, y2, color='red')
plt.plot(x3, y3, color='green')
plt.plot(x4, y4, color='yellow')
plt.plot(x5, y5, color='pink')
plt.plot(x6, y6, color='black')

plt.title('Ages of Nobel prize winners in all categories in decades during 1900-2020', fontdict = font1, loc = 'center')
plt.legend(['Chemistry', 'Physics', 'Medicine', 'Economics', 'Literature', 'Peace'], bbox_to_anchor=(1,1), loc="upper left")
plt.ylabel('Ages', fontdict = font2, loc = 'center')
plt.xlabel('Decades', fontdict = font2, loc = 'center')
plt.xticks(rotation=90)
#ax.grid(True)
plt.show()
```

# Pie chart and donut chart (percentage of categories; percentage of male, female and organizations of Nobel prise winners)

```python
import plotly.graph_objects as go
```

```python
orgg=pd.read_csv('nobel_prize_winner.csv')
org=orgg[['id', 'gender', 'category']]
org.head(2)
```

```python
# Pie charm
chemistry=org.loc[org['category'] == 'chemistry'].count()[0]
physics=org.loc[org['category'] == 'physics'].count()[0]
medicine=org.loc[org['category'] == 'medicine'].count()[0]
economics=org.loc[org['category'] == 'economics'].count()[0]
literature=org.loc[org['category'] == 'literature'].count()[0]
peace=org.loc[org['category'] == 'peace'].count()[0]
colors1 = ['red', 'lightgreen', 'yellow', 'blue', 'lightpink', 'cyan']
labels = ['chemistry', 'physics', 'medicine', 'economics', 'literature', 'peace']
fig = plt.figure(figsize=(10,10)) 
plt.rcParams.update({'font.size': 20, 'font.family': 'monospace'})
plt.pie([chemistry, physics, medicine, economics, literature, peace], labels=labels, colors=colors1, autopct='%.2f %%', wedgeprops={'edgecolor': 'black'})
plt.title('Individual winners in the fields', color='blue')
plt.show()
```

```python
s=org.gender
counts=s.value_counts()
percent100=s.value_counts(normalize=True).mul(100).round(1).astype(str)# + '%'
ff=pd.DataFrame({'counts': counts, 'percent100': percent100})
ff
```
